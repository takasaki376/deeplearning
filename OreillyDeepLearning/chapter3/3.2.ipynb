# 3.2.2 ステップ関数の実装

```python
import numpy as np
import matplotlib.pyplot as plt

def step_func(x):
    return np.array(x > 0 ,dtype=np.int)
```

```python
x = np.arange(-5.0 , 5.0, 0.1)
y1 = step_func(x)
plt.ylim(-0.1, 1.1)
plt.plot(x , y1)
plt.show()
```

# 3.2.4 シグモイド関数の実装

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.arange(-5.0 , 5.0, 0.1)
y2 = sigmoid(x)
plt.ylim(-0.1, 1.1)
plt.plot(x , y2)
plt.show()
```

# 3.2.5 シグモイド関数とステップ関数の比較

```python
plt.ylim(-0.1, 1.1)
plt.plot(x , y1,linestyle = "dashed")
plt.plot(x , y2)
plt.show()
```

# 3.2.7 ReLU関数

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0 , x)
```

```python
x = np.arange(-5.0 , 5.0, 0.1)
y3 = relu(x)
plt.ylim(-0.1, 5.1)
plt.plot(x , y3)
plt.show()
```
